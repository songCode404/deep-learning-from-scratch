#### 1. 퍼셉트론이란?
- 다수의 신호를 입력으로 받아서 하나의 신호를 출력하는 것

#### 2. 임계값과 가중치의 역할
##### 2-1. 임계값(theta)이란?
뉴런에서 보내온 신호가 활성화되기 위한 기준값
- 뉴런에서 보내온 신호의 총합이 임계값보다 작으면 활성화 X
- 뉴런에서 보내온 신호의 총합이 임계값보다 크면 활성화 O

##### 2-2. 가중치(weight)란?
전류에서 말하는 저항과 같은 것으로, **가중치가 크면 클수록 더 강한 신호를 흘려보냄.**
- 즉, 가중치가 크면 더 중요한 신호라는 뜻. 더 반영을 많이 하게 됨

#### 3. 학습이란?
- 퍼셉트론에서 우리는 AND, OR, NAND를 구현할 때 적절한 가중치와 임계값을 직접 설정해줘야 했음
- 머신러닝을 통한 **학습은 이런 매개변수의 값을 정하는 작업을 컴퓨터가 자동으로 하게 하는것**
- 즉 **학습 = 적절한 매개변수 값을 정하는 작업**



```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1

print(AND(0,0)) #0을 출력
print(AND(1,0)) #0을 출력
print(AND(0,1)) #0을 출력
print(AND(1,1)) #1을 출력
```

#### 4. 가중치와 편향 도입
##### 4-1. 편향이란?
**뉴런이 얼마나 쉽게 활성화하느냐를 조정하는 매개변수**
<-> 가중치 : 입력 신호가 결과에 주는 영향력(중요도)을 조절하는 매개변수

- 편향(-b) = 임계값(theta) 

```python
import numpy as np

#편향을 추가한 AND 게이트
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7 

    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1
    
print(AND(0,0)) #0을 출력
print(AND(1,0)) #0을 출력
print(AND(0,1)) #0을 출력
print(AND(1,1)) #1을 출력
```

```python
#편향을 추가한 NAND 게이트
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7 

    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1
    
print(NAND(0,0)) #1을 출력
print(NAND(1,0)) #1을 출력
print(NAND(0,1)) #1을 출력
print(NAND(1,1)) #0을 출력
```

```python
#편향을 추가한 OR 게이트
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2

    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1
    
print(OR(0,0)) #0을 출력
print(OR(1,0)) #1을 출력
print(OR(0,1)) #1을 출력
print(OR(1,1)) #1을 출력
```

#### 📍5. 퍼셉트론의 한계
**직선 하나로 나눈 영역만 표현할 수 있음**
**즉, 곡선으로 나눠야하는 영역은 표시할 수 없음**
- 배타적 논리합을 표현하는 논리 회로인 **XOR 게이트를 표현할 수 없음**
    - 배타적 논리합이란? 
        - x1과 x2중 한쪽이 1일 때만 1을 출력하는 논리 회로


#### 💡6. 해결 방안
**층을 쌓아서 "다층 퍼셉트론"을 만들기!!!!!!**

- XOR 게이트는 NAND 게이트, OR 게이트, AND 게이트로 만들 수 있음.


##### 📍**층을 하나 늘려서 단층 퍼셉트론으로는 표현하지 못한 것을 구현할 수 있음**

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)

    return y

print(XOR(0, 0)) #0을 출력
print(XOR(1, 0)) #1을 출력
print(XOR(0, 1)) #1을 출력
print(XOR(1, 1)) #0을 출력
```
